```python
# Calculate loss function and then minimize the error after Nth iterations by adjusting the parameters.

import numpy as np

# Use the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the loss function (Mean Squared Error)
def loss(predicted, actual):
    return np.mean((predicted - actual) ** 2)

# Neural network architecture
input_size = 2  # For AND, OR, NAND, NOR gates (adjust for NOT gate)
hidden_size = 3
output_size = 1

# Initialize weights and biases
np.random.seed(42)  # Set seed for reproducibility
W1 = np.random.randn(input_size, hidden_size)
b1 = np.random.randn(hidden_size)
W2 = np.random.randn(hidden_size, output_size)
b2 = np.random.randn(output_size)

# Define the input and target outputs for each gate

# AND Gate
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
target = np.array([[0], [0], [0], [1]])

# OR Gate
# x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
# target = np.array([[0], [1], [1], [1]])

# NOT Gate
# Note: NOT gate is a single-input gate, so we redefine the input and target
# input_size = 1
# x = np.array([[0], [1]])
# target = np.array([[1], [0]])

# NAND Gate
# x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
# target = np.array([[1], [1], [1], [0]])

# NOR Gate
# x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
# target = np.array([[1], [0], [0], [0]])

# Variables to track the best loss and associated weights and biases
best_loss = float('inf')
best_W1, best_b1, best_W2, best_b2 = None, None, None, None

# Training loop
for iteration in range(150):
    # Forward pass
    hidden_layer_output = np.dot(x, W1) + b1
    output = sigmoid(np.dot(hidden_layer_output, W2) + b2)

    # Save the final output
    final_output = output

    # Calculate the loss
    current_loss = loss(output, target)

    # Check if we have found a better loss
    if current_loss < best_loss:
        print(f'New weights found, iteration: {iteration}, loss: {current_loss}')
        best_loss = current_loss
        best_W1, best_b1 = W1.copy(), b1.copy()
        best_W2, best_b2 = W2.copy(), b2.copy()
    else:
        # Revert to the previous best weights and biases
        W1, b1 = best_W1.copy(), best_b1.copy()
        W2, b2 = best_W2.copy(), best_b2.copy()

    # Update weights and biases with small random values
    W1 += 0.09 * np.random.randn(input_size, hidden_size)
    b1 += 0.09 * np.random.randn(hidden_size)
    W2 += 0.09 * np.random.randn(hidden_size, output_size)
    b2 += 0.09 * np.random.randn(output_size)

# Print the final output
print("Final Output:")
print(final_output)
```

```python
# Solve AND gate problem using neural network

import numpy as np

# Define the inputs and corresponding labels for the 2-input AND gate
inputs = np.array([[0, 0],
                   [0, 1],
                   [1, 0],
                   [1, 1]])

labels = np.array([0, 0, 0, 1])

# Define the weights and biases for the neural network
weights = np.array([0.5, 0.5])
bias = -0.7

# Define the activation function (step function)
def step_function(x):
    return 0 if x <= 0 else 1

# Perform the forward pass to compute the outputs
outputs = []
for input_row in inputs:
    weighted_sum = np.dot(input_row, weights) + bias
    output = step_function(weighted_sum)
    outputs.append(output)

# Compare the predicted outputs with the actual labels
correct_predictions = np.sum(outputs == labels)

# Print the results
print("Inputs:", inputs)
print("Labels (Actual Outputs):", labels)
print("Predicted Outputs:", outputs)
print("Correct Predictions:", correct_predictions)
```
